### Matrix A

```python
def find_apis(block):
    return re.findall('(invoke-.*?)\\n', block)
```

```python
def smali2apis(row):
    smalis = '\n'.join(row.dropna())
    return set(find_apis(smalis))
```

```python
apis = smalis.apply(smali2apis, axis = 1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
```

```python
A = pd.DataFrame(mlb.fit_transform(apis),columns=mlb.classes_, index=apis.index)
```

### Matrix B

```python
def find_blocks(smali):
    return re.findall( '\.method([\S\s]*?)\.end method', smali)
```

```python
def smali2blocks(row):
    smalis = '\n'.join(row.dropna())
    return list(set(find_blocks(smalis)))
```

```python
blocks = smalis.apply(smali2blocks, axis = 1).explode().reset_index().iloc[:,1]
```

```python
block_api = blocks.dropna().apply(find_apis)
```

```python
block_api.apply(pd.Series)
```

```python
apis_df = pd.DataFrame.from_dict(dict(zip(block_api.index, block_api.values))).T
```
